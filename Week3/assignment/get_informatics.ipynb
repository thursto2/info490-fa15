- Print every number from 1 to 100.
- If the number is a multiple of 3, print "info" instead.
- If the number is a multiple of 5, print "matics" instead.
- If the number is a multiple of both 3 and 5, print "informatics" instead.

Note the spelling of "informatics". There's an R in informatics.
